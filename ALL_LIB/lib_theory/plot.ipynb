#   Scatter   Plot

- **Use   when:**

   - You   want   to   visualize   relationship   between   two   numerical   features

   - Good   for   seeing   clusters,   patterns,   outliers

- **Example:**
   - Age   vs   Salary,   Height   vs   Weight

- **Best   use:**

   -   Regression   visualization

   -   Clustering   visualization

   -   Outlier   detection

   -   Decision   boundaries

   -   PCA/t-SNE   visualization

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,5,4,5]

plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

```python
from sklearn.datasets import make_classification, make_blobs
from sklearn.decomposition import PCA

# Generate synthetic data
X, y = make_classification(n_samples=1000, n_features=20, n_informative=2, 
                           n_redundant=10, n_classes=3, n_clusters_per_class=1)

# PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(12, 5))

# Basic scatter plot
plt.subplot(1, 2, 1)
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', 
                      alpha=0.6, s=50, edgecolor='k', linewidth=0.5)
plt.colorbar(scatter, label='Class')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization')

# t-SNE for non-linear visualization
from sklearn.manifold import TSNE
X_tsne = TSNE(n_components=2, random_state=42).fit_transform(X)

plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='Set3', 
            alpha=0.7, s=30, edgecolor='k', linewidth=0.3)
plt.title('t-SNE Visualization')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.tight_layout()
```

# Pair Plot (Seaborn)

- ## Use when:

# Confusion Matrix Plot
- ## Use when:

  - You are evaluating a classification model

  - Helps see where model is making mistakes

- ## Example:
   - True Positive, False Positive, False Negative, True Negative

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_true = [0,1,0,1,0,1]
y_pred = [0,1,0,0,1,1]

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix")
plt.show()
```

# ROC Curve

- **Use when:**

  - Evaluating binary classifier

  - To check performance at all thresholds

- ### Shows:

  - True Positive Rate (TPR)

  - False Positive Rate (FPR)

```python
from sklearn.metrics import roc_curve
import numpy as np

# Fake predictions
y_true = [0,1,0,1,0,1]
y_prob = [0.2,0.8,0.3,0.7,0.1,0.6]

fpr, tpr, _ = roc_curve(y_true, y_prob)
plt.plot(fpr, tpr)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.show()
```

# Precision-Recall Curve

- **Use when:**

  - The dataset is imbalanced

  - Better than ROC in such cases

```python
from sklearn.metrics import precision_recall_curve

precision, recall, _ = precision_recall_curve(y_true, y_prob)
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```
