# Threshold Interpretation

What is a Threshold?
A threshold is a cutoff value used to convert model outputs (probabilities, scores) into discrete class predictions.



# Binary Classification Example
probabilities = model.predict_proba(X)[:, 1]  # [0.23, 0.67, 0.89, 0.45]
threshold = 0.5
predictions = (probabilities > threshold).astype(int)  # [0, 1, 1, 0]


### Why Threshold Matters?
Standard Threshold (0.5)
Assumes equal cost of false positives and false negatives

Optimal when class distribution is balanced

When to Adjust Threshold:
Medical diagnosis: Lower threshold (0.3) to catch more diseases

Spam detection: Higher threshold (0.9) to avoid blocking important emails

Fraud detection: Very low threshold (0.1) to catch all suspicious activity

### Threshold & Evaluation Metrics
Confusion Matrix Metrics:
``` bash
                       Actual
                 Positive   Negative
Predicted  Pos    TP         FP
           Neg    FN         TN

 ```
### Threshold affects:

Precision = TP / (TP + FP) - Higher threshold increases

Recall = TP / (TP + FN) - Lower threshold increases

F1-Score = Harmonic mean of precision & recall

Specificity = TN / (TN + FP) - Higher threshold increases
